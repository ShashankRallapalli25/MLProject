### 3.7 Exploring Data

```python
print("categories in 'gender' variable: ", df['gender'].unique(), end=" ")
print("\ncategories in 'race_ethnicity' variable: ", df['race_ethnicity'].unique())
print("\ncategories in 'parental level of education' variable: ", df['parental_level_of_education'].unique())
print("\ncategories in 'lunch' variable: ", df['lunch'].unique())
print("\ncategories in 'test_preparation_course' variable: ", df['test_preparation_course'].unique())
```

```python
#define numerical and categorical columns
numeric_features = [feature for feature in df.columns if df[feature].dtype !='O']
categorical_features = [feature for feature in df.columns if df[feature].dtype =='O']

print("We have {} numeric features :{} " .format(len(numeric_features), numeric_features))
print("We have {} categorical features :{} " .format(len(categorical_features), categorical_features))
```

```python
#adding total and avg columns
df['total_score']= df['math_score']+df['reading_score']+df['writing_score']
df['average_score']=df['total_score']/3
df.head(2)
```

```python
reading_full = df[df['reading_score']==100]['average_score'].count()
writing_full = df[df['writing_score']==100]['average_score'].count()
math_full = df[df['math_score']==100]['average_score'].count()

print(f"count of students with 100% reading score: {reading_full}")
print(f"count of students with 100% writing score: {writing_full}")
print(f"count of students with 100% math score: {math_full}")
```

```python
reading_less = df[df['reading_score']<=20]['average_score'].count()
writing_less = df[df['writing_score']<=20]['average_score'].count()
math_less = df[df['math_score']<=20]['average_score'].count()

print(f"count of students with 20% or less reading score: {reading_less}")
print(f"count of students with 20% or less writing score: {writing_less}")
print(f"count of students with 20% or less math score: {math_less}")
```

#####  Insights
 - From above values we get students have performed the worst in Maths 
 - Best performance is in reading section

```python
df_gen = df[['gender','math_score','reading_score','writing_score','total_score','average_score']]
gender_group = df_gen.groupby('gender').mean()
gender_group.head()
```

```python
plt.figure(figsize=(5, 4))

X = ['Total Average','Math Average']


female_scores = [gender_group['average_score'][0], gender_group['math_score'][0]]
male_scores = [gender_group['average_score'][1], gender_group['math_score'][1]]

X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, male_scores, 0.4, label = 'Male')
plt.bar(X_axis + 0.2, female_scores, 0.4, label = 'Female')
  
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total average v/s Math average marks of both the genders", fontweight='bold')
plt.legend()
plt.show()
```

#### Insights 
- On an average females have a better overall score than men.
- whereas males have scored higher in Maths.
